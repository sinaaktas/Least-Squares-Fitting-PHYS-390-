```python
from math import log10,floor
import numpy as np
%matplotlib notebook
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

test=100000  #numbers generated
N_1 = np.random.uniform(0,1,test)
N_2 = np.random.uniform(0,1,test)
N_3 = np.random.uniform(0,1,test)
N_4 = np.random.uniform(0,1,test)
N_5 = np.random.uniform(0,1,test)

#using just round for significant figures not work well because it count zeros
def round_sig(x, sig=6):     #this function calculate after zeros
    return round(x, sig-int(floor(log10(abs(x))))-1)

t =(N_1+N_2+N_3+N_4+N_5)/5 #array of mean of the 5 random generated numbers
print('Experimental Variance:',"            ", round_sig(np.var(t)),"       ","(sig=6)")
print('Mathematically calculated Variance:', 0.0166667,"       ","(sig=6)")
print()
print('ExperimentalMean',"                  ",round_sig(np.mean(t)),"        ","(sig=6)")
print('Mathematically calculated Mean:',"   ", '0.500000'"         ","(sig=6)")


plt.hist(t,bins=100)    
plt.axvline(x=np.mean(t)-np.std(t), ls = "--", color='green', alpha=0.7,label="Stdev")
plt.axvline(x=np.mean(t)+np.std(t), ls = "--", color='green', alpha=0.7)
plt.axvline(x=np.mean(t), ls = "--", color='red', alpha=0.7,label="Mean")
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)


plt.show()
```

```python
import math
import numpy as np
%matplotlib notebook
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

tt=np.round(t,decimals=2,out=None) #it is for the proper histogram range

x=np.array([i for i in range(0,101)])/100 #0 to 1, 101 point in fit graph
y=np.exp(-((x-np.mean(t))**2/(2*np.std(t)**2)))/(np.std(t)*np.sqrt(2*np.pi)) #normalized gaussian function

plt.hist(t,bins=101,rwidth=1,density=True,range=[min(tt)-0.005,max(tt)+0.005])    
plt.axvline(x=np.mean(t)-np.std(t), ls = "--", color='green', alpha=0.7,label="Stdev")
plt.axvline(x=np.mean(t)+np.std(t), ls = "--", color='green', alpha=0.7)
plt.axvline(x=np.mean(t), ls = "--", color='red', alpha=0.7,label="Mean")
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
plt.plot(x,y,'k',alpha=0.7)
plt.show()
```

### Part 3
(Not Normalized)

\begin{equation}
CARUANA’S \;\;ALGORITHM
\end{equation}
\begin{equation}
Fit\; function\;y(x)\;=\;A\cdot e^{-(x-\mu)^2/2\sigma}
\end{equation}
\begin{equation}
A\;=\;e^{a-b^2/4c}
\end{equation}
\begin{equation}
\mu\;=\;\frac{-b}{2c}\quad , \quad \sigma=\sqrt{\frac{-1}{2c}}
\end{equation}
\begin{equation}
only\; unknown\; is \;"a"
\end{equation}
\begin{equation}
To\; find\; "a" \; I \; will \; use \; Least \; Square \; Method
\end{equation}

```python
import math
import numpy as np
%matplotlib notebook
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

tt=np.round(t,decimals=2,out=None) #we have all avaraged of 5 numbers in 2 digit, then every number of times of frequency will equal to bin length
c=-1/(2*np.var(t)) #c and b in Caruana's Algorithm
b=-2*c*np.mean(t)
x=np.array([x for x in range(0,101)])/100 # it is for the x line of histogram 
                                          #(101 x values and I will set 101 bins)

min_diff=10e15
best_a=0
a=0 # I tried it in high range, for running the code fast I'll start it from 0
for h in range(0,500): #It search best "a" between 0 and 1
    diff=0
    a+=0.002
    y_1=(np.exp(a-((b**2)/(4*c))))*np.exp(-((x-np.mean(t))**2/(2*np.std(t)**2)))# fit function as array
    for i in range(0,101):
        diff+=(y_1[i]-np.count_nonzero(tt == i*0.01))**2 #it calculate sums of delta square of 101 point between histogram and function
    if diff<min_diff:
        min_diff=diff 
        best_a=round(a,3)
        

y=np.exp(best_a-((b**2)/(4*c)))*np.exp(-((x-np.mean(t))**2/(2*np.std(t)**2))) #final version of our fit function
plt.hist(t,bins=101,range=[min(tt)-0.005,max(tt)+0.005])

plt.axvline(x=np.mean(t)-np.std(t), ls = "--", color='green', alpha=0.7,label="Stdev")
plt.axvline(x=np.mean(t)+np.std(t), ls = "--", color='green', alpha=0.7)
plt.axvline(x=np.mean(t), ls = "--", color='red', alpha=0.7,label="Mean")
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
plt.plot(x,y,'k',alpha=0.7)
plt.show()
print('Best a value is:',best_a)
print(y[0],y[100])
```
